# ✨ single neural network XOR 연산 ✨
: a1x1 + a2x2 + b

```python
import tensorflow as tf
import numpy as np
```

```python
learning_rate = 0.1
np.random.seed(0)
tf.set_random_seed(0)

x_data = np.array([[0,0], [0,1], [1,0], [1,1]])
y_data = np.array([[0], [1], [1], [0]])

X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

W1 = tf.Variable(tf.random_normal([2, 2]), name = 'weight1')   ## 두 개의 값 들어가서 두 개의 값 나옴
b1 = tf.Variable(tf.random_normal([2]), name = 'bias1')
layer1 = tf.sigmoid(tf.matmul(X, W1) + b1)               ## 히든 레이어 (은닉층)

W2 = tf.Variable(tf.random_normal([2, 1]), name = 'weight2')   ## 두 개의 값 들어가서 한 개의 값 나옴
b2 = tf.Variable(tf.random_normal([1]), name = 'bias2')
hypothesis = tf.sigmoid(tf.matmul(layer1, W2) + b2)


cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

predicted = tf.cast(hypothesis > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        sess.run(train, feed_dict={X:x_data, Y:y_data})
        
        if step % 100 == 0:
            print("step =", step, "cost =", sess.run(cost, feed_dict={X:x_data, Y:y_data}),
                 "W2 =", sess.run(W2))
            
    # Accuracy
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X:x_data, Y:y_data})
    print("\nHypothesis : \n", h, "\nCorrect : \n", c, "\nAccuracy : \n", a)
```

# ✨ ReLU 함수 ✨
: 입력값이 0보다 작을 때는 아예 non-activate 로 꺼버리고, 0보다 클 때는 그 값을 그대로 반환  
=> 0보다 작을 때는 미분값이 0이고, 0보다 클 때는 미분값이 1이다      
=> 기울기 소실 문제 해결!!!👍

=> 여러가지 활성화 함수 (시그모이드, 렐루, 하이퍼볼릭 탄젠트, 소프트플러스)

```python
learning_rate = 0.1
np.random.seed(0)      ## 랜덤시드는 초기값만 고정됨!! => 그 이후는 매번 미세하게 바뀐다!!
tf.set_random_seed(0)

x_data = np.array([[0,0], [0,1], [1,0], [1,1]])
y_data = np.array([[0], [1], [1], [0]])

X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

W1 = tf.Variable(tf.random_normal([2, 5]), name = 'weight1')   ## 두 개의 값 들어가서 다섯 개의 값 나옴
b1 = tf.Variable(tf.random_normal([5]), name = 'bias1')
layer1 = tf.nn.relu(tf.matmul(X, W1) + b1)               ## 히든 레이어 (은닉층)

W2 = tf.Variable(tf.random_normal([5, 5]), name = 'weight2')   ## 다섯 개의 값 들어가서 다섯 개의 값 나옴
b2 = tf.Variable(tf.random_normal([5]), name = 'bias2')
layer2 = tf.nn.relu(tf.matmul(layer1, W2) + b2)          ## 히든 레이어 (은닉층)

W3 = tf.Variable(tf.random_normal([5, 5]), name = 'weight3')   ## 다섯 개의 값 들어가서 다섯 개의 값 나옴
b3 = tf.Variable(tf.random_normal([5]), name = 'bias3')
layer3 = tf.nn.relu(tf.matmul(layer2, W3) + b3)          ## 히든 레이어 (은닉층)

W4 = tf.Variable(tf.random_normal([5, 5]), name = 'weight4')   ## 다섯 개의 값 들어가서 다섯 개의 값 나옴
b4 = tf.Variable(tf.random_normal([5]), name = 'bias4')
layer4 = tf.nn.relu(tf.matmul(layer3, W4) + b4)          ## 히든 레이어 (은닉층)

W5 = tf.Variable(tf.random_normal([5, 1]), name = 'weight5')   ## 다섯 개의 값 들어가서 한 개의 값 나옴
b5 = tf.Variable(tf.random_normal([1]), name = 'bias5')
hypothesis = tf.sigmoid(tf.matmul(layer4, W5) + b5)

## ⭐⭐마지막 부분(hypothesis)은 0과 1 사이로 표현을 하기 위해 sigmoid를 사용!!⭐⭐


cost = -tf.reduce_mean(Y * tf.log(hypothesis) + (1 - Y) * tf.log(1 - hypothesis))
train = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)

predicted = tf.cast(hypothesis > 0.5, dtype = tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, Y), dtype = tf.float32))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for step in range(10001):
        sess.run(train, feed_dict={X:x_data, Y:y_data})
        
        if step % 100 == 0:
            print("step =", step, "cost =", sess.run(cost, feed_dict={X:x_data, Y:y_data}),
                 "W2 =", sess.run(W2))
            
    # Accuracy
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={X:x_data, Y:y_data})
    print("\nHypothesis : \n", h, "\nCorrect : \n", c, "\nAccuracy : \n", a)
```

```python
# sigmoid 함수 정의
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
def z(x, w, b):
    return w * x * b            # z = wx + b
def H(x):
    return sigmoid(w * x + b)   # H(x) = sigmoid(z) = sigmoid(wx + b)
```

```python
# 오차제곱평균법 사용시

# w_list = np.linspace(-1.2, 1.6, 50)
w_list = np.linspace(-20, 30, 51)
b = 0
x = np.linspace(-5, 10, 51)
y = np.append(np.zeros(30), np.ones(21))

Err = []
for w_value in w_list:
    z = w_value * x + b
    Err.append(sum((sigmoid(z)-y)**2) / len(y))  ## w에 따른 오차제곱평균
    
print(Err)

plt.title("w와 Err의 그래프", fontdict = font1)
plt.xlabel("w", fontdict = font2)
plt.ylabel("Err", fontdict = font3)
plt.grid(True)     ## 그리드(격자)가 보이게
plt.plot(w_list, Err)   ## x축이 z, y축이 sigmoid(z)
plt.show()
```

## sigmoid 함수에 대한 cost 함수

```python
w_list = np.linspace(-7, 3, 101)
b = 0
x = np.linspace(-5, 10, 51)
y = np.append(np.zeros(40), np.ones(11))

cost = []
for w_value in w_list:
    z = w_value * x + b
    h = sigmoid(z)
    c = - y * np.log(h) - (1 - y) * np.log(1 - h)
    cost.append(sum(c) / len(c))   ## 각 기울기 값 w에 대한 cost의 총합을 list에 기록
    
plt.title("기울기 w에 대한 cost 그래프", fontdict = font1)
plt.xlabel("w", fontdict = font2)
plt.ylabel("cost", fontdict = font3)
plt.grid(True)     ## 그리드(격자)가 보이게
plt.plot(w_list, cost)   ## x축이 z, y축이 sigmoid(z)
plt.show()
```
