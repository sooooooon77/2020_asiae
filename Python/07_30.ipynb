# 방탄소년단 뉴스 리스트
- title
- url
- 심화 (description)

```python
resp = requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8")     ## 위의 코드와 같은 코드
resp
```

```python
resp.url
```

```python
soup = BeautifulSoup(resp.text)
type(soup)
```

### 방법1 
★ 권장하는 방법 ★

```python
ul_tags = soup.find('ul', class_ = 'type01')

news_li = ul_tags.find_all('li', recursive = False)
print(len(news_li))
news_li
```

```python
result_list = []

for news in news_li:
    title_tag = news.find('a', class_ = '_sp_each_title')
    title = title_tag.text
    url = title_tag.get('href')
    
    desc = news.find_all('dd')
    desc = desc[1].text           ### 위에서 찾은 모든 dd 중에서 두 번째 dd의 text (0,1,2,,,,,)
    
    result_list.append({
        'title':title,
        'url':url,
        'desc':desc
    })

print(len(result_list))
result_list
```

### 방법2 
- select 이용

```python
print(len(soup.select('ul.type01>li>dl>dt>a')))
titles = soup.select('ul.type01>li>dl>dt>a')
```

```python
## css selector

print(len(soup.select('ul.type01>li>dl>dd:nth-child(3)')))    ## dl 의 세번째 자식이 dd 이면 가져오라 라는 뜻!!
descs = soup.select('ul.type01>li>dl>dd:nth-child(3)')
```

```python
result_list = []

for a_tag, desc_tag in zip(titles, descs):
    result_list.append({
        'title':a_tag.text,
        'url':a_tag.get('href'),
        'desc':desc_tag.text
    })

print(len(result_list))
result_list
```

# 블로그 데이터 가져오기
- title
- url
- 심화 (description)

```python
resp = requests.get("https://search.naver.com/search.naver?date_from=&date_option=0&date_to=&dup_remove=1&nso=&post_blogurl=&post_blogurl_without=&query=%EB%B0%A9%ED%83%84%EC%86%8C%EB%85%84%EB%8B%A8&sm=tab_pge&srchby=all&st=sim&where=post&start=1")     ## 위의 코드와 같은 코드
resp
```

```python
def crawl_blog(query, page_num = 2):

    url = "https://search.naver.com/search.naver"


    result_list = []

    for i in range(page_num):
        params = {
        'where' : 'post',
        'query' : query,
        'start' : i*10 + 1
        }

        resp = requests.get(url, params = params)
        resp

        soup = BeautifulSoup(resp.text)


        news_li = soup.find_all('li', class_ = 'sh_blog_top')


        for news in news_li:
            title_tag = news.find('a', class_ = 'sh_blog_title')
            title = title_tag.text
            blog_url = title_tag.get('href')

            desc = news.find_all('dd')
            desc = desc[1].text           ### 위에서 찾은 모든 dd 중에서 두 번째 dd의 text (0,1,2,,,,,)

            result_list.append({
                'title':title,
                'url':blog_url,
                'desc':desc

            })

    return result_list
```

```python
blog_search_list = crawl_blog("트와이스", 3)
print(len(blog_search_list))
blog_search_list
```

```python
base_url = "https://finance.naver.com/item/sise_day.nhn"
params = {
    'code' : '005930',
}

resp = requests.get(base_url, params = params)
resp
```

```python
soup = BeautifulSoup(resp.text)
type(soup)
```

```python
tr_tags = soup.select('table.type2 tr')
len(tr_tags)     ## header 들 6 개 포함 16개!
tr_tags
```

```python
th_tags = tr_tags[0].find_all('th')

title_list = []
for th in th_tags:
    title_list.append(th.text)

print(title_list)
```

```python
result_list = []

for tr_tag in tr_tags: 
    td_tags = tr_tag.find_all('td')
    
    if len(td_tags) < 7:     ## 길이가 7 보다 작은건 버림
        continue
    
    data_list = []
    
    for td in td_tags:
        data_list.append(td.text.strip())  ## strip : 공백제거
        
    result_list.append(
        dict(zip(title_list, data_list))
    )
    
print(result_list)
```

```python
def get_day_sise(code, page_num):
    
    base_url = "https://finance.naver.com/item/sise_day.nhn"
    
    result_list = []
    
    for page in range(1, page_num+1):
        params = {
            'code' : '005930',
            'page' : page
        }

        resp = requests.get(base_url, params = params)
        soup = BeautifulSoup(resp.text)
        
        tr_tags = soup.select('table.type2 tr')
        
        th_tags = tr_tags[0].find_all('th')

        title_list = []
        for th in th_tags:
            title_list.append(th.text)

        

        for tr_tag in tr_tags: 
            td_tags = tr_tag.find_all('td')

            if len(td_tags) < 7:     ## 길이가 7 보다 작은건 버림
                continue

            data_list = []

            for td in td_tags:
                data_list.append(td.text.strip())  ## strip : 공백제거

            result_list.append(
                dict(zip(title_list, data_list))
            )

    return result_list
```

```python
sise = get_day_sise('005930', 5)
sise
```
