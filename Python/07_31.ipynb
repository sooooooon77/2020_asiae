```python
import requests
from bs4 import BeautifulSoup
```

```python
resp = requests.get("https://news.naver.com/main/list.nhn?mode=LS2D&mid=shm&sid1=105&sid2=731")     ## 위의 코드와 같은 코드
resp
```

```python
soup = BeautifulSoup(resp.text)
type(soup)
```

```python
ul_tags = soup.find('ul', class_ = 'type06_headline')


# titles = soup.select('ul.type01>li>dl>dt>a')
# descs = soup.select('ul.type01>li>dl>dd:nth-child(3)')

news_li = soup.select('ul.type06_headline>li>dl>dt:nth-child(2)')
print(len(news_li))
news_li
```

```python
result_list = []

for news in news_li:
    title_tag = news.find('a')
    title = title_tag.text.strip()
    url = title_tag.get('href')
    
    
    
    resp = requests.get(url)  
    soup = BeautifulSoup(resp.text)
    div_tags = soup.find('div', class_ = '_article_body_contents')
    
    content = div_tags.text.strip()
    
    
    ul_tags = soup.find('ul', class_ = 'u_cbox_list')
    div_tags2 = ul_tags('div', class_ = 'u_cbox_text_wrap')
    
    for comment in div_tags2:
        comment_tag = comment.find('span', class_ = 'u_cbox_contents')
        comments = comment_tag.text
      
    
    
    result_list.append({
        'title':title,
        'url':url,
        'content':content,
        'comment':comments
    })

print(len(result_list))
result_list
```
