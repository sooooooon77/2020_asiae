# merge

```python
ktx_df['rank']='A'
ktx_df.loc[:2013,'rank']='B'
ktx_df.loc[[2014,2016],'rank']='C'
ktx_df
```

```python
rank_df=pd.DataFrame(
    {
        'rank':['A','B','C','D'],
        'price':[100,10,1,0.1]
    }
)
rank_df
```

```python
ktx_df.merge(rank_df,on='rank') #rank가 똑같은 기준으로
```

```python
ktx_df.merge(rank_df,on='rank',how='outer') #합집합
```

```python
ktx_df.merge(rank_df,on='rank',how='left') #왼쪽 기준으로 넣음(ktx_df 기준)
```

```python
ktx_df.merge(rank_df,on='rank',how='right') #오른쪽 기준으로 넣음(rank_df 기준)
```

# where

```python
def categorizing_by_5000(data):
    data=data.where(lambda x: x>=5000,0) #5000보다 크거나 같으면 냅둬,아니면 0
    data=data.where(lambda x: x<5000,1)
    return data
```

```python
ktx_df.iloc[:,:-2].apply(categorizing_by_5000)
```
